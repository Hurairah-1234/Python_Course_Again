```python
# imploting the graph
import seaborn as sns
# sns.set_theme(style="")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("titanic")

# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df, x="age", y="fare",
    col_wrap=2, palette="muted", ci=None,
    col="class", hue="class", height=4,
)
```

```python
# scatter plot
sns.set_theme(style="whitegrid")
sns.scatterplot(x="age", y="fare", data=df)
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="age", y="fare",
                hue="class", size="class",
                palette="ch:r=-.2,d=.3_r",
                sizes=(10, 200), linewidth=0,data=df, ax=ax)
```

```python
# line plot
# sns.set_theme(style="darkgrid")
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
# fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="age", y="fare", style="class", data=df)
             
```

```python
# dis plot

sns.set_theme(style="darkgrid")
# df = sns.load_dataset("penguins")
sns.displot(x="age", col="class",row="sex",binwidth=3, height=3, facet_kws=dict(margin_titles=True), data=df)
    
```

```python
#rel plot
# import seaborn as sns
sns.set_theme(style="darkgrid")

# dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
# palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(data=df, x="age", y="fare",hue="class",kind="line", size_order=["T1", "T2"],height=5, aspect=.75, facet_kws=dict(sharex=False),)
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

```python
# rel plot
sns.set_theme(style="darkgrid")
sns.relplot(data=df, x="age", y="survived",hue="sex",kind="line")
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

df_1 = sns.load_dataset("titanic")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_1, kind="bar",
    x="class", y="pclass", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
```

```python
# box plot
from tkinter.ttk import Style
import seaborn as sns
# sns.set_theme(style="ticks", palette="pastel")


# Load the example tips dataset
df = sns.load_dataset("titanic")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="survived", y="fare", hue="class",data=df)
            

sns.despine(offset=10, trim=True)
```

```python
# violin plot
# import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the example tips dataset
df = sns.load_dataset("titanic")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=df, x="survived", y="fare", hue="class")
               
```
